```python
from nltk.tag import pos_tag
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer

```

```python
def clean_grammer(value:list,
                  stemmer_=SnowballStemmer('english',ignore_stopwords=False),
                  lemmatizer_=WordNetLemmatizer()):
  clean_value=list()
  for word,pos in pos_tag(value):
    if stemmer_:word=stemmer_.stem(word)
    if pos.startswith('J'): word=lemmatizer_.lemmatize(word,'r')
    elif pos.startswith('R'): word=lemmatizer_.lemmatize(word,'s')
    elif pos.startswith('V'): word=lemmatizer_.lemmatize(word,'v')
    elif pos.startswith('N'): word=lemmatizer_.lemmatize(word,'n')
    else: word=lemmatizer_.lemmatize(word,'n')
    clean_value.append(word)
  return clean_value

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def clean(value:str,tags2remove:list=['http[\S]+','@[\S]+','&amp[\S]+'],
          tag2convert:dict={'positive ':[':)',';',':-)',':p',':D'],
                            'negative ':[':(',';(',':-(']},
          cleaner:str='[^a-zA-Z ]',
          stopwords:list=stopwords.words(fileids='english'),
          minimum_word_length:int=3,call_clean_grammer:bool=True,
          clean_grammer_method=clean_grammer):
  if isinstance(value,(list,tuple)):
    value=[clean(sentc) for sentc in value]
    return value
  for tag in tags2remove:
    value= re.sub(tag,'',value)
  for tag_name,tags in tag2convert.items():
    for tag in tags:
      value=value.replace(tag,tag_name)
  value=re.sub(cleaner,'',value)
  value=value.casefold()
  value=word_tokenize(value)
  value=[word for word in value if word not in stopwords]
  value=[word for word in value if not len(word)<minimum_word_length]
  if call_clean_grammer: return clean_grammer_method(value)
  else: return value





```

```python
def tokenizeSentenceGen(data:list)->dict:
  for sentence in data:
    yield dict([(word,True) for word in sentence])
def dataGenerator(data:list):
  for sentence in data:
    for token in sentence:
      yield token
```

```python
for index in random_indexs:
  print('Tweet number:',index,':',positive_tweets[index])
```

```python
cleaned_tokenized_positive_tweets=clean(positive_tweets)
for index in random_indexs:
  print('Tweet number:',index,':',cleaned_tokenized_positive_tweets[index])
```

```python
for index in random_indexs:
  print('Tweet number:',index,':',negative_tweets[index])
```

```python
cleaned_tokenized_negative_tweets=clean(negative_tweets)
for index in random_indexs:
  print('Tweet number:',index,':',cleaned_tokenized_negative_tweets[index])
```

```python
positiveGenerator=dataGenerator(cleaned_tokenized_positive_tweets)
print(f'typeof positive_word_gen_ ->{positiveGenerator}:',
      type(positiveGenerator))
from nltk.probability import FreqDist
positive_words_frequency=FreqDist(positiveGenerator)
print(positive_words_frequency)
print('Frequency of 10 most common elements [positive]:\n',
      positive_words_frequency.most_common(10),sep='')
type(positive_words_frequency)
```

```python
negativeGenerator=dataGenerator(cleaned_tokenized_negative_tweets)
print(f'typeof negative_word_gen_ ->{negativeGenerator}:',
      type(negativeGenerator))
from nltk.probability import FreqDist
negative_words_frequency=FreqDist(negativeGenerator)
print(negative_words_frequency)
print('Frequency of 10 most common elements [negative]:\n',
      negative_words_frequency.most_common(10),sep='')
type(negative_words_frequency)
```

```python
positive_words4Data=tokenizeSentenceGen(cleaned_tokenized_positive_tweets)
negative_words4Data=tokenizeSentenceGen(cleaned_tokenized_negative_tweets)

print('Sample:',next(tokenizeSentenceGen(cleaned_tokenized_positive_tweets)))

positive_dataset=[(tweet_dict,'Positive') for tweet_dict in positive_words4Data]
negative_dataset=[(tweet_dict,'Negative') for tweet_dict in negative_words4Data]

dataset=positive_dataset+negative_dataset
print('Size of dataset:',len(dataset))

from sklearn.model_selection import train_test_split

train_set,test_set = train_test_split(dataset,test_size=0.10)
len(train_set),len(test_set)
```

```python
from nltk import classify, NaiveBayesClassifier
classifier_ = NaiveBayesClassifier.train(train_set)
print('Accuracy_score:',round(classify.accuracy(classifier_,test_set)*100,3))
```
